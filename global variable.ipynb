```python
def func(x):
  print("local x",x)
  x=2
  print("changed local",x)
x=50
func(x)
print("x is still",x)
```

```python
def func():
  y=2
  print("changed y to",y)
y=50
func()
print(y)
```

```python
def func():
  global y
  y=2
  print("changed y to",y)
y=50
func()
print(y)
```
