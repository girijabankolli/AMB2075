```python
import pandas as pd
data={
    "farm":["farm A","farm B","farm C","farm D"],
      "CROP":["wheat","rice","maize","barley"],
      "AREA":[10,15,8,12],
      "production":[30,60,45,25],
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
df.head(2)
```

```python
df.tail()
```

```python
df.tail(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
df["yield_ton_per_ha"]=df["production"]/df["AREA"]
print(df)
```

```python
print(df.iloc[1:4:2])
```

```python
print(df.iloc[[1,3]])
```

```python
print(df.iloc[[1,3],[0,1,2]])
```

```python
print(df['CROP']=="wheat")
print()
wheat_farm=df[df["CROP"]=="wheat"]
print(wheat_farm)
```

```python
high_yield=df[df["yield_ton_per_ha"]>3]
print(high_yield)
```

```python
sort=df.sort_values(by="yield_ton_per_ha",ascending=False)
print(sort)
```

```python
avg_yield=df.groupby("CROP")["yield_ton_per_ha"].mean()
print(avg_yield)
```

```python
sum_yield=df.groupby("CROP")["yield_ton_per_ha"].sum()
print(sum_yield)
```
