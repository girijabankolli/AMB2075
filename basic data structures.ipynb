```python
num1 = 15
num2 = 4

print(f"Addition: {num1} + {num2} = {num1 + num2}")
print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
print(f"Multiplication: {num1} * {num2} = {num1 * num2}")
print(f"Division (float): {num1} / {num2} = {num1 / num2}")
print(f"Division (integer): {num1} // {num2} = {num1 // num2}")
print(f"Modulus (remainder): {num1} % {num2} = {num1 % num2}")
print(f"Exponentiation: {num1} ** {num2} = {num1 ** num2}")
```

```python
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Test cases
print(f"10 is {check_even_odd(10)}")
print(f"7 is {check_even_odd(7)}")
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Test cases
print(f"Is 17 prime? {is_prime(17)}")
print(f"Is 4 prime? {is_prime(4)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 2 prime? {is_prime(2)}")
```
