```python
l1=[10,20,30,40,50,60]
n=[]
for i in l1:
  n.append(i+2)
print("new list",n)
```

```python
l1=[10,20,30,40,50,60]
n=[i+2 for i in l1]
print("new list",n)
```

```python
l1=[10,20,30,40,50,60]
n=[]
for i in l1:
  if i%4==0:
    n.append(i)
    print("numbers divisible by 4",n)
```

```python
l1=[10,20,30,40,50,60]
n=[i for i in l1 if i%4==0]
print("new list",n)
```
