```python
print("hey....welcome to FARMING GAME✌🎁(❁´◡`❁)")
print("lets start the game 🌾3️⃣..2️⃣..1️⃣...")
print("choose the option")
print("1.plant seeds 🌱")
print("2.water crops 💧")
print("3.harvest crops🌾")
print("4.exit")
crops=0
points=0

while(True):
  ch=int(input("enter your choice 😀"))
  if ch==1:
    crops +=1
    print("you planted 1 crop..🥰🌾")
  elif ch==2:
    if crops >0:
     print("watered..💧🌾")
    else:
     print("no crops to water..😥😥")
  elif ch==3:
     if crops >=1:
      print("harvesting is done.you got 10 points..🤑🤩")
      points=crops*10
     else:
      print("no crops to harvest..😥")
  elif ch==4:
    print("end game")
    print("points you earned..🤩🤩",points)
    break
  else:
    print("invalid choic..😥")


```
