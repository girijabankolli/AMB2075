```python
crops=["rice","wheat","maize"]
water_need=[200,400,500]
max_water=max(water_need)
index=water_need.index(max_water)
print("crop with highest water requirement:",crops[index])
```

```python
crops=["maize","rice","wheat"]
y=[200,400,500]
max(y)
average_yield=sum(y)/len(crops)
print("high yield crops:")
for i in range(len(crops)):
 if (y[i]>average_yield):
  print(crops[i])
```

```python
crops=["rice","maize","wheat"]
fertilizer=["urea","dap","compost"]
for c,n in zip(crops,fertilizer):
  print( c ,"fertilizer reccomendation is",n)

```
