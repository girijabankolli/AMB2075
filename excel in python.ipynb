```python
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"],color="black")
plt.xlabel("Rainfall_mm")
plt.ylabel("Production_Tons",color="red")
plt.title("Rainfall vs Production",color="red")
plt.show()
```

```python
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print(yearly_production)
```

```python
plt.figure(figsize=(5,5))
plt.bar(yearly_production.index,yearly_production.values,color="skyblue")
plt.xlabel("Year")
plt.ylabel("Total Production")
plt.title("yearly production")
plt.xticks()
```

```python
Wheat=df[df["Crop"]=="Wheat"]
Rice=df[df["Crop"]=="Rice"]
plt.plot(Wheat["Year"],Wheat["Production_Tons"],label="Wheat")
plt.plot(Rice["Year"],Rice["Production_Tons"],label="Rice")
```

```python
Wheat=df[df["Crop"]=="Wheat"]
Rice=df[df["Crop"]=="Rice"]
plt.plot(Wheat["Year"],Wheat["Production_Tons"],label="Wheat")
plt.plot(Rice["Year"],Rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
plt.ylabel("Production_Tons")
plt.title("crop production over years")
plt.legend()
```

```python
dat_2022=df[df["Year"]==2022]
print(dat_2022)
plt.figure(figsize=(3,2))
plt.bar(dat_2022["Crop"],dat_2022["Production_Tons"],color=["orange","black"])
plt.xlabel("Crop")
plt.ylabel("Production_Tons")
plt.title("Crop production in 2022")
```
