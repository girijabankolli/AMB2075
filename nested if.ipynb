```python

# Assume a student's mark out of 100
student_mark = 78

# Define the passing mark
passing_mark = 50

print(f"Evaluating student with mark: {student_mark}")

# Outer if statement: Check if the student passed
if student_mark >= passing_mark:
    print("Congratulations! You have passed.")

    # Nested if-elif-else statement: Determine the specific grade
    if student_mark >= 90:
        print("Grade: A+ (Excellent!)")
    elif student_mark >= 80:
        print("Grade: A (Very Good)")
    elif student_mark >= 70:
        print("Grade: B (Good)")
    else:
        print("Grade: C (Satisfactory)")

    # Another nested if for an additional condition, e.g., distinction
    if student_mark > 85:
        print("You achieved a distinction!")

else:
    print("Unfortunately, you have not passed.")
    print("Grade: F (Fail)")

print("Mark evaluation complete.")
```
