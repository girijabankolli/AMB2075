```python
import numpy as np
crop_yield=np.array([
    [12.5,2.4,6.5,5.5,8.7],
    [12.5,2.4,6.5,5.5,8.7],
    [2.3,5.8,8.9,9.0,9.0],
    [2.3,5.8,8.9,9.0,9.0] ,
    [7.3,5.8,8.9,9.0,9.0]

])
print(crop_yield)
print(crop_yield.shape)
print(crop_yield.dtype)
print(crop_yield.size)
print(crop_yield.ndim)
```

```python
total_yield=np.sum(crop_yield)
avg_farm=np.mean(crop_yield,axis=1)
avg_yield_per_season=np.mean(crop_yield,axis=0)
print(total_yield)
print(avg_farm)
print(avg_yield_per_season)
max_yield=np.max(crop_yield)
min_yield=np.min(crop_yield)
```

```python
best_farm=np.argmax(avg_farm)
print(best_farm+1)
```
