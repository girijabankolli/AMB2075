```python
eqp=["tractor","plough","harvester","seeder","sprayer","thresher"]
cost=[850000,25000,1200000,60000,18000,95000]
for e,c in zip (eqp,cost):
  print(eqp)
```

```python
max_amt=max(zip(cost,eqp))
print(max_amt)
```

```python
print("expensive equipment is",max_amt[1])
```

```python
min_amt=min(zip(cost,eqp))
print(min_amt)
```

```python
print("least expensive equipment is",min_amt[1])
```

```python
d=(max_amt[0]-min_amt[0])
print(d)
```

```python
avg=sum(cost)/len(cost)
print(avg)
```

```python
for i in range(len(cost)):
  if (cost[i])==avg:
    print(eqp[i])
  else:
    print("eqp cost is not =avg cost")
```

```python
for i in range(len(cost)):
  if cost[i]>avg:
    print(eqp[i])
```

```python
for i in range(len(cost)):
  if cost[i]<50000:
    print(len(eqp[i]))
```

```python
for i in range(len(cost)):
  if cost[i]<=1000000:
    print("eligible for gov subsidy",eqp[i])
```
